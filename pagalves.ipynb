# Pixel manipulation

```python
doggy_pixels = doggy.getdata()
for pixel in doggy_pixels:
    print(pixel)
```

```python
changed_doggy_pixels = []
for pixel in doggy_pixels:
    if pixel[0] >= 150 and pixel[1] >= 150 and pixel[2] >= 180:
        new_pixel = pixel[0] - 50, pixel[1] - 50, pixel[2]
    else:
        new_pixel = pixel
    changed_doggy_pixels.append(new_pixel)
changed_doggy_pixels
```

```python
doggy.putdata(changed_doggy_pixels)
doggy
```
