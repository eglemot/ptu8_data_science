```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.shape
```

```python
sns.set_style('whitegrid')
```

```python
sns.displot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'], bins = 15, kde=True)
```

```python
sns.scatterplot(tips, x='total_bill', y='tip', size='size')
```

```python
sns.jointplot(tips, x='total_bill', y='tip', kind='hex', cmap='rocket_r', color=sns.color_palette('rocket')[1])
```

```python
sns.jointplot(tips, x='total_bill', y='tip', hue='size', palette='coolwarm')
```

```python
sns.jointplot(tips, x='total_bill', y='tip', kind='reg')
```

```python
sns.jointplot(tips, x='total_bill', y='tip', kind='kde', cmap='coolwarm')
```

```python
sns.scatterplot(tips, x='total_bill', y='tip', size='size', hue='sex')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,diag_kind='list', hue='sex')
```

```python
sns.barplot(tips, x='sex', y='total_bill')
```

```python
sns.barplot(tips, x='sex', y='total_bill', estimator='sum')
```

```python
sns.barplot(tips, x='sex', y='total_bill',hue='size', estimator='sum')
```

```python
sns.barplot(tips, x='sex', y='total_bill',hue='day', estimator='sum')
```

```python
sns.countplot(tips, x='smoker')
```

```python
sns.boxplot(tips, x='smoker', y='total_bill')
```

```python
koreliacija = tips.corr(numeric_only=True)
koreliacija
```

```python
sns.heatmap(koreliacija, annot=True)
```
