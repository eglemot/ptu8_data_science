# Iterator/Generator


```python
def poriniai(iki):
    count = 2
    while count <= iki:
        yield count
        count += 2
```

```python
skaiciuojam = poriniai(99)
```

```python
print(next(skaiciuojam))
```

```python
for skaicius in skaiciuojam:
    print(skaicius)
```

```python
def fib(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b
        
```

```python
ats = fib(45)
print(list(ats))
```

```python
kodas = "1149"
def lauziam_koda():
    skaicius = 0
    while skaicius < 10000:
        spejimas = f'{skaicius:04}'
        if spejimas == kodas:
            print(f'{spejimas} yra tai, ko jus ieskote')
            break
        yield spejimas
        skaicius += 1
```

```python
bandom = lauziam_koda()
while True:
    try:
        print(next(bandom))
    except StopIteration:
        break
```

# DEKORATORIAI



```python
def stri_decorator(fn):
    def wrapper(*args, **kwargs):
        for a in args:
            if type(a) != str:
                return "text must be a string"
        return fn(*args, **kwargs)
    return wrapper
```

```python
@stri_decorator
def return_string(*args):
    return args
```

```python
return_string("")
```

```python
def less_than_two(fn):
    def wrapper(*args, **kwargs):
        if len(args) > 2:
            return "Enter less than 2 parameters"
        return fn(*args)
    return wrapper
```

```python
@stri_decorator
@less_than_two
def any_function(*args):
    return args
```

```python
any_function("stringas", "stringiukas")
```

```python
from time import sleep
from functools import wraps

def delay(seconds):
    def delay(fn):
        @wraps(fn)
        def wrapper(*args, **kwargs):
            print(f"function will be delayed for {seconds} seconds. ")
            sleep(seconds)
            return fn(*args, **kwargs)
        return wrapper
    return delay
```

```python
@delay(2.5)
@stri_decorator
@less_than_two
def any_function(*args):
    return args
```

```python
any_function("stringas", "stringiukas")
```

### how long does it takes 

```python
from time import time
def performance(fn):
    def wrapper(*args, **kwargs):
        t1 = time()
        result = fn(*args, **kwargs)
        t2 = time()
        diff = t2 - t1
        # print(f"It took {diff}")
        return result, diff
    return wrapper
```

```python
import requests  # importuojame requests
from time import time  # importuojame time modulį

start_time = time()  # fiksuojame starto laiką
r = requests.get('http://www.cnn.com')  # sukuriame http užklausą
print(r.status_code)  # spausdiname status code (200 = OK, 404 = Not Found, ir t.t. galima pasiguglinti http status codes)
end_time = time()  # fiksuojame pabaigos laiką
print(end_time - start_time)  # atspausdiname laiką, per kurį gaovme atsakymą
```
